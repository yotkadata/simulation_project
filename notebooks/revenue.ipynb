```python
import pandas as pd
```

```python
supermarket = pd.read_csv('../simulation/data_prep.csv', parse_dates = ['timestamp'],index_col=[0])
supermarket
```

```python
# data of the each customer total in supermarket
total_time = pd.read_csv('../simulation/total_timespent.csv')
total_time
```

```python
# Create a DataFrame with the revenue per minute for each section from course material
revenue_data = {'location': ['fruit', 'spices', 'dairy', 'drinks'],
                'revenue_per_minute': [4, 3, 5, 6]}
revenue_df = pd.DataFrame(revenue_data)
revenue_df
```

```python
# Calculate the estimated revenue for each customer in each section
customer_revenue = pd.merge(supermarket, revenue_df, on='location', how='left')
customer_revenue['estimated_revenue'] = customer_revenue['revenue_per_minute'] * total_time['time_spent']
customer_revenue
```

```python
# Calculate the total estimated revenue for each section
location_revenue = customer_revenue.groupby('location')['estimated_revenue'].sum().reset_index()
location_revenue
```

```python
# Find the most profitable section
most_profitable_section = location_revenue.loc[location_revenue['estimated_revenue'].idxmax(), 'location']
most_profitable_section
```

```python

# Display the most profitable section
print(f"The most profitable section is: {most_profitable_section}")
```
