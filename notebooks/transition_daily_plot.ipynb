```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mon = pd.read_csv("monday_tp.csv", index_col = 0)
tue = pd.read_csv("tuesday_tp.csv", index_col = 0)
wed = pd.read_csv("wednesday_tp.csv", index_col = 0)
thu = pd.read_csv("thursday_tp.csv", index_col = 0)
fri = pd.read_csv("friday_tp.csv", index_col = 0)
```

```python
mon["day"] = "monday"
tue["day"] = "tuesday"
wed["day"] = "wednesday"
thu["day"] = "thursday"
fri["day"] = "friday"
```

```python
df = pd.concat([mon, tue, wed, thu, fri], axis=0)
df
```

```python
filtered_df = df[(df.index == "dairy") & (df['day'].isin(["monday", "tuesday", "wednesday", 'thursday', "friday"]))]
filtered_df
```

```python
# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10, 8))

sns.barplot(data = filtered_df, x="day", y='dairy', ax=axes[0, 0])
#if you want to add titles for subplots: axes[0, 0].set_title('Plot 1')
sns.barplot(data = filtered_df, x="day", y='drinks', ax=axes[0, 1])
sns.barplot(data = filtered_df, x="day", y='fruit', ax=axes[1, 0])
sns.barplot(data = filtered_df, x="day", y='spices', ax=axes[1, 1])
sns.barplot(data = filtered_df, x="day", y='checkout', ax=axes[2, 1])

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(25, 5))

sns.barplot(data = filtered_df, x="day", y='dairy', ax=axes[0])
sns.barplot(data = filtered_df, x="day", y='drinks', ax=axes[1])
sns.barplot(data = filtered_df, x="day", y='fruit', ax=axes[2])
sns.barplot(data = filtered_df, x="day", y='spices', ax=axes[3])
sns.barplot(data = filtered_df, x="day", y='checkout', ax=axes[4])

plt.title("Dairy to Other Sections")
plt.tight_layout()
plt.show()
```
